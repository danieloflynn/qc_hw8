```python
from qiskit import QuantumCircuit, transpile, assemble
from qiskit.quantum_info import Operator
from qiskit.circuit.library import CCZGate
from qiskit_aer import Aer
import numpy as np
```

```python
backend = Aer.get_backend('qasm_simulator')
backend_options = {'method': 'statevector'}
```

```python
def run_simulation(quantum_circuit, backend_options=backend_options, noise_model=None, name='psi_1'):
    job = backend.run(quantum_circuit, run_options=backend_options, noise_model=noise_model)
    result    = job.result()
    psi_out = result.data(0)[name]
    psi_out_img = psi_out.draw('latex', prefix='|\\psi_{out}> = ')
    display(psi_out_img)
```

```python
ccz = CCZGate()
qc = QuantumCircuit(3)

qc.h(0)
qc.h(1)
qc.h(2)

# Oracle
qc.x(0)
qc.x(1)
qc.x(2)
qc.append(ccz, [0, 1, 2])

qc.x(0)
qc.x(1)
qc.x(2)

# Diffusion
qc.h(0)
qc.h(1)
qc.h(2)
qc.x(0)
qc.x(1)
qc.x(2)

qc.append(ccz, [0, 1, 2])

qc.x(0)
qc.x(1)
qc.x(2)

qc.h(0)
qc.h(1)
qc.h(2)

qc.save_statevector(label='psi_out')
qc.draw("mpl")
```

```python
run_simulation(qc, name='psi_out')
```
